```python
bids=[[1,2,5,0],
     [2,1,4,2],
     [3,5,4,6]]
totalShares=3
```

```python
def getUnalottedUsers(bids,totalShares):
    shares = 0
    unalotted = []
    alotted=[]
    for b in bids:
        shares = shares+b[1] 
    #if shares are not enough for everyone
    if totalShares >= shares: 
        return unalotted  

    bids.sort(key = lambda x: (-x[2],x[3]))  
    
#     print('here')
#     print(bids)

    samePrice=[]
    for i in range(len(bids)-1):
        if (bids[i][2]==bids[i+1][2]):
            samePrice.append(bids[i][0])
            if(bids[i+1][0] not in samePrice):
                samePrice.append(bids[i+1][0])            

    for b in bids:
        if (totalShares <= 0): 
            unalotted.append(b[0])
        else:
            if(b[0] in samePrice):
                totalShares = totalShares-1
            else:
                totalShares = totalShares - b[1]
    
    unalotted=sorted(unalotted)
    return unalotted
```

```python
unalotted=getUnalottedUsers(bids,totalShares)
print(unalotted)
```
