```python
after=[[2,5],
      [7,17]]
```

```python
def findBeforeMatrix(after):
    n = len(after)
    m = len(after[0])
    
    
    before=[]
    for i in range(n):
        x=[]
        for j in range(m):
            x.append(0)
        before.append(x)
            
    for i in range(n):
        for j in range(m):
            #for every value in before
            s=0
            for x in range(i+1):
                for y in range(j+1):
                    if(x!=i or y!=j):
                        s=s+before[x][y]
            before[i][j]=after[i][j]-s
    
    return before
    
before = findBeforeMatrix(after)            
print(before)
```
