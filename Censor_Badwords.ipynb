```python
badwords='crikey lame jerk dummy bollocks'
message='In this handout (lame), we will refer to this as the "bollocks" controlling idea, because it controls what happens in the rest of the dummy paragraph.'
```

```python
bad=badwords.split(' ')

#print(bad)

punct= [' ',',','!','(',')','.','"']
word=''
sentence=''
#print('starting')
for i in message:
    if (i not in punct):
        word= word + i
    if (i in punct):
        #print(word)
        #print('encoutered = ',i)
        if word in bad:
            sentence=sentence + str('*'*len(word)) + i
        else:
            sentence=sentence + word + i
        word=''

print(sentence)
```
